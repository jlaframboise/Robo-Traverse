```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
```

```python
folder = "~/Desktop/Design Teams/QMIND/Experiment1Data"
file = "AllData.csv"

df = pd.read_csv(os.path.join(folder, file))
```

```python
df = df.rename(columns={'Unnamed: 0': 'Seq'})
df = df.set_index('Seq')
df.head(10)
```

```python
print(df.isnull().sum().sum())
df.shape
```

```python
# combined_labels = df['Speed'].map(str) + "_" + df['Terrain']
# combined_labels
```

```python
le = preprocessing.LabelEncoder()
```

```python
X = df.drop(columns=['Speed', 'Terrain'])
# y = le.fit_transform(combined_labels)
y = le.fit_transform(df['Terrain'])
```

```python
scaler = StandardScaler()

scaledFeatures = scaler.fit_transform(X.values)

X = pd.DataFrame(scaledFeatures, columns = X.columns.tolist())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
